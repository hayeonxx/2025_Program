###for 활용

- for문은 정해진 횟수만큼 반복해서 문장을 실행
- 셀수 있는 것: 범위, 문자열, 리스트, 딕셔너리 등
- 일반적으로 범위를 가장 많이 사용하며, 범위는 range()함수 사용



```
for 변수 in 셀 수 있는 것:
  문장
```



- range() 함수: range(시작값, 종료값, 증가값)
  - 시작값으로 시작해서 증가값 만큼 증가하며 종료값 바로 전까지 값을 생성
  - 증가값을 입력하지 않으면 증가값은 1
  - 시작값을 입력하지 않으면 기본 시작값은 0

- 문자열 출력을 반복하는 프로그램

````python
for i in range(3):
  print("안녕!")
````

- for문 값이 변하는 변수

````python
for a in range(2,5):
  print(a)
````

- range 함수에서 증가값이 음수인 경우

````python
for a in range(10,1,-3):
  print(a)
````

- 정수의 합을 구하는 프로그램

````python
sum = 1+2+3+4+5
print(sum)
````

````python
sum = 0
for a in range(1,6):
  sum += a
  print(sum)
````

###전 직원의 월급 총액 구하기

- 직급별 월급이 다름
- 전체 직원수 변수 : employee_num
- 월급 총액 변수 : total
- 2급 이상: 본봉 + 상여금(30%)
- 2급 미만: 본봉 + 상여금(20%)

````python
employee_num = 3
total = 0

for i in range(employee_num):
  rank = int(input('직급: '))
  basic_salary = int(input('본봉: '))
  if rank >= 2:
    bonus = basic_salary*0.3
  else:
    bonus = basic_salary*0.2
  salary = basic_salary + bonus
  total += salary
print('월급총액: ',total)
````

###경진대회 평가 점수 구하기

- 심사위원들의 평가 점수 합: sum
- 최고 점수 : max_score
- 조건
  - 심사위원 5명
  - 공정성을 위해 최고 점수를 뺀 나머지 점수 평균

````python
judge_num = 5
sum = 0
max_score = 0

for i in range(judge_num):
  score = int(input(f'심사위원 {i+1} 평가 점수: '))
  sum += score
  if score > max_score:
    max_score = score

sum -= max_score
print('작품 평가 점수 : ', sum/(judge_num-1))
````

###적정 심박수 구하기

- 카이보넨 공식: 나이와 안정시 심박수를 이용해 운동에 필요한 적정 심박수를 계산하는 공식
- 목표 심박수 = ((((220-나이)-안정 시 심박수)*강도)+안정 시 심박수)

````python
age = int(input('나이 : '))
resting_HR =int(input('안정 시 심박수: '))
for a in range(40, 81, 10):
  intensity = a / 100
  target_HR = (((220-age)-resting_HR)*intensity)+resting_HR
  print(f'강도 {a}% 목표 심박수 : {target_HR}bpm')
````

##While문

###While문의 활용
- 조건식이 참인 동안 문장을 반복해서 실행
- 거짓이면 종료



```
while 조건식:
  문장
```



````python
a = 1
while a<=9:
  print(f'{a}회 반복')
  a += 1
````

- 정수의 합

````python
sum = 0
a = 1
while a <= 10:
  sum += a
  a += 1
print(sum)
````

- 입력한 값에 따라 다르게 동작하는 프로그램
  - 이름 : 문자열로 입력 받음
  - 나이 : 숫자로 입력 받음
  - 프로그램 종료 : 종료 문자열을 입력으로 받음
- 입력 받은 내용이 숫자 형식인지 판별하는 함수 사용 : isdigit()
  - 문자열이 숫자인지 판별, 숫자 형식이면 True 아니면 False
  - object.method()
- 반복문을 빠져 나올 때는 break 키워드 사용

````python
while True:
  answer = input('이름 or 나이 or 종료: ')
  if answer == '종료':
    print('종료합니다.')
    break
  elif answer.isdigit():
    print(f'{answer}세 군요')
  else:
    print(f'{answer}님 반갑습니다.')
````

###5만원 모으기
- 하루에 100원부터 100원씩 추가하여 모을 경우 몇일이 걸리는지 구하는 프로그램
  - sum : 저금한 총액
  - money : 일자별로 저금하는 금액
  - day는 저금한 일수

````python
sum = 0
money = 100
day = 0

while sum < 50000:
  sum += money
  day += 1
  money += 100

print(day)
````

###자리 수 합 구하기
- 정수 각 자릿수의 합 구하기
  -ex : 234 -> 2+3+4 = 9

````python
num = int(input('정수 : '))
sum = 0

while num>0:
  sum += num%10
  num = num//10

print(sum)
````

###사용자가 생각한 수 맞추기

````python
print('1~100중 숫자 하나를 정하세요')
print('이 숫자를 맞혀보겠습니다.')

low = 1
high = 100
times = 0

while True:
  num = (low+high)//2
  answer = input(f'{num}인가요? OK/UP/DOWN: ')
  times += 1
  if answer == 'OK':
    break
  elif answer == 'UP':
    low = num + 1
  elif answer == 'DOWN':
    high = num - 1
  else:
    print('잘못 입력했습니다.')

print(f'정답입니다. {times}회 만에 맞췄습니다.')
````
