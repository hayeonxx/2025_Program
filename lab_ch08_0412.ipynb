###선 그리기
- 라이브러리 추가 및 plt 이름 사용
  - import matplotlib.pyplot as plt
  - plt.plot([x],y, [옵션])
  - x를 생량하면 x축의 값은 0부터 1씩 증가하는 리스트가됨
- plot()를 사용하려면 matplotlib를 불러와야 함
- 라이브러리 이름이 길 경우 as keyword를 이용하여 이름을 줄여 줌
- plot()함수로 그린 그래프는 show()함수를 이용하여 표시

````python
import matplotlib.pyplot as plt

year = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
gdp = [3079.9,3250.1,3398.8,3574,3678.2,3721.8,3744,4003.6,4165.5]

plt.plot(year, gdp)
plt.show()
````

###응용: Dictionary의 keys와 values값을 list로 변환하여 그래프 그리기

````python
year = {2014:3079.9,2015:3250.1,2016:3398.8,2017:3574,2018:3678.2,2019:3721.8,2020:3744,2021:4003.6,2022:4165.5}

year_list = list(year.keys())
gdp_list = list(year.values())

plt.plot(year_list, gdp_list)
plt.show()
````

###히스트그램 그리기
- 도수 분포를 직사각형 기둥으로 나타내는 그래프
- hist()를 이용
  - 기본 10개의 구간으로 나누어 그림
  - 막대의 색상
  - 구간의 개수 등 설정가능

###LAB

####LAB 혈액형 빈도수 시각화하기
- 혈액형을 입력받고 빈도수를 시각적으로 비교
- 종료가 입력될 때까지 혈액형을 입력받아 딕셔너리로 처리

````python
import matplotlib.pyplot as plt

blood = {'A':0, 'B':0, 'O':0, 'AB':0}

while True:
  s = input('혈액형(A,B,O,AB) 또는 종료: ')
  if s=='종료':
    break
  elif s in blood:
    blood[s] += 1
  else:
    print('잘못 입력했습니다.')

plt.figure(figsize=(5,3))
plt.bar(blood.keys(), blood.values(), width=0.6)
plt.xlabel('blood type')
plt.ylabel('frequency')
plt.show()
````

####LAB 난수를 이용하여 히스토그램 그리기

- 정규분포
  - random 라이브러리의 normalvariate()를 이용: 정규분포를 따르는 난수 생성
  - random.normalvariate(평균, 표준편차)

````python
import matplotlib.pyplot as plt
import random

x = [random.normalvariate(0,1) for _ in range(1000)] # i 변수 대신 기호(_)를 사용하기도 함

plt.figure(figsize=(5,3))
plt.ylabel('frequency')
plt.hist(x, bins=20, alpha=0.5)
plt.show()
````

####이미지 제목 설정
- 이미지 축 눈금 없애는 plt.axis('off')를 해지해야 함
- x,y축에 값이 표시되지 않게 눈금의 위치, 눈금의 이름을 설정하는 xticks함수를 이용해서 x,y축 값을 빈 리스트로 설정
- xlabel()를 이용하여 리스트 항목 표시

````python
#import matplotlib.pyplot as plt

cat = plt.imread('/content/cat.jpg')
dog = plt.imread('/content/dog.jpg')

imgs = [cat, dog]
labels = ['cat', 'dog']
plt.figure(figsize=(5,3))
for i in range(len(imgs)):
  plt.subplot(1,2,i+1)
  plt.imshow(imgs[i])
  plt.xticks([])
  plt.yticks([])
  plt.xlabel(labels[i])
plt.show()
````
