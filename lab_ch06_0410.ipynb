###문자열 인덱싱
- 문자열의 각 문자에 접근할 떄 문자 위치인 index를 사용
- 문자에 접근하는 것은 indexing이라함
- 문자의 길이를 알려주는 함수, len()

```python
s = 'abc'
for i in range(len(s)):
  print(s[i], end = ' \n')
```

####인덱스 없이 문자에 접근

```python
s = 'abc'
for a in s:
  print(a, end=' ')
```

###문자열 슬라이싱
- 슬라이싱(slicing):[:]를 이용해서 문자열의 일부 문자를 추출
- 문자열[시작 인덱스:종료 인덱스]
- 종료 인덱스 직전 문자까지 추출

```python
s = 'abcde'
print(s[2:4])
```

```python
print(s[:3])
print(s[3:])
```

###문자열 거꾸로 출력

```python
s = input('문자열: ')
for i in range(len(s)-1, -1, -1):
  print(s[i], end = '')
```

####회문 판별
- 회문(palindrome):문자열을 거꾸로 뒤집어도 원래 문자열과 같은 문자
- s2에 문자열 저장 후 s1 == s2 비교하여 회문 판별

```python
s1 = input('문자열: ')
s2 = ''

for i in range(len(s1)-1, -1, -1):
  s2 = s2 + s1[i]

if s1 == s2:
  print('회문')
else:
  print('회문 아님')
```

###문자열 구성 파악하기
- 사용자가 입력한 문자열이 조건에 맞는지 판별할 수 있음

####아이디 입력
- 입력받은 아이디가 5자리 이상
- 알파벳과 숫자로만 이루어질 때까지 입력받기를 반복

```python
while True:
  id = input('아이디: ')
  if len(id)<5:
    print('길이가 짧습니다.')
  elif id.isalnum() != True:
    print('아이디는 알파벳과 숫자로 이루어져야 합니다.')
  else:
    break
print(f'아이디는 {id}입니다.')
```

####형 변환

```python
s = input('키: ')

if s.find('cm') != -1:
  height = float(s.rstrip('cm'))
else:
  height = float(s)
print(height)
```

####문자 함수

- 문자 함수

```python
print(chr(65))
print(ord('A'))
```

- 소문자 2개를 입력받아 알파벳 순서를 기준으로 얼마나 떨어져 있는지 구하는 프로그램

```python
c1 = input('알파벳 소문자: ')
c2 = input('알파벳 소문자: ')

if c1 > c2:
  diff = ord(c1) - ord(c2)
else:
  diff = ord(c2) - ord(c1)
print('떨어진 정도:',diff)
```

####대소문자 상호 변환
- 대문자는 소문자로, 소문자는 대문자로
  - 사용자가 입력한 문자열 대문자는 소문자로 소문자는 대문자로 변환

```python
s1 = input('문자열 : ')
s2 = ''

for ch in s1:
  if ch.islower():
    s2 += ch.upper()
  elif ch.isupper():
    s2 += ch.lower()
  else:
    s2 += ch
print(s2)
```
