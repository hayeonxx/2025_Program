####원하는 열 추려내기
- iso_code:국제 표준 국가 코드(USA,KOR)
- location:국가 이름
- date:일자
- total_cases:전체 확진자 수
- population:인구

```python
#원하는 열 리스트 만들기
selected_columns = ['iso_code','location','date','total_cases','population']
```

```python
#선택한 열로 새로운 데이터프레임 생성
revised_df = raw_df[selected_columns]
```

```python
#생성한 데이터프레임 확인
revised_df .head()
```
