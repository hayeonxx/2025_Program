#Chap.05 리스트(List)

- 여러 개의 데이터를 저장하는 공간

##리스트의 개요

###리스트의 개념

- 리스트(list): 여러 개의 데이터를 저장하는 공간
  - ex) 값 5개를 저장하려면 변수 5개를 만듬
  - 리스트 하나를 만들어 5개의 값을 한꺼번에 저장할 수 있음

<리스트의 개념이 없을 경우>
- 5개의 가격을 저장

````python
price1 = 1500
price2 = 3200
price3 = 2500
price4 = 4000
price5 = 5500
````

- 5개의 가격이 5% 인상되었다면

````python
price1 = price1 + price1*0.05
price2 = price2 + price2*0.05
price3 = price3 + price3*0.05
price4 = price4 + price4*0.05
price5 = price5 + price5*0.05
````

````python
print(price1)
print(price2)
print(price3)
print(price4)
print(price5)
````

<리스트를 사용하는 경우>
- 대괄호([]) 안에 데이터를 콤마(,)로 구분하여 입력
- 각 데이터를 항목(item) 또는 요소(element)라 함

- 인덱스(index): 대괄호 안에 표시되는 숫자, 시작은 0부터

````python
prices = [1500, 3200, 2500, 4000, 5500]
print(prices)
````

````python
i = 0
for i in range(len(prices)):
  prices[i] = prices[i]+prices[i]*0.05

print(prices)
````

##리스트 생성
- 리스트 이름 = [항목1, 항목2, ...]

###값을 입력하여 리스트 생성

````python
menus = ['짜장면', '짬뽕', '탕수육', '군만두']
print(menus)
````

###리스트에 값 추가하기

````python
menus = []
menus.append('짜장면')
menus.append('짬뽕')
menus.append('탕수육')
menus.append('군만두')
print(menus)
````

### 반복문으로 리스트 생성
- 1부터 10까지의 정수를 세제곱한 수를 cubes에 저장

````python
cubes = []

for i in range(1,11):
  cubes.append(i**3)

print(cubes)
````

###리스트 내포(list comprehension)

````python
cubes = [x**3 for x in range(1,11)]
print(cubes)
````

- [식 for 변수 in 반복 가능한 것 if 조건식]

````python
sequence = []
for x in range(1,19):
  if x % 7 == 0:
    sequence.append(x)

print(sequence)
````

````python
sequence = [x for x in range(1,19) if x % 7 == 0]
print(sequence)
````

- 리스트 안 성적이 70 이상이면 합격 아니면 불합격

````python
score = [71,80,60,90,65]
result = []

for a in score:
  if a>= 70:
    result.append('합격')
  else:
    result.append('불합격')

print(result)
````

````python
score = [71,80,60,90,65]
result = ['합격' if a >= 70 else '불합격' for a in score]
print(result)
````

##리스트 관련 함수

###항목 삭제하기
- pop()
- del 키워드
- remove()
- clear()

#### pop()
- 리스트.pop(인덱스)
- 인덱스를 생략하면 마지막 항목을 삭제하고 변환

````python
pets = ['dog', 'cat', 'hamster', 'hedgehog', 'parrot', 'lizard']
select1 = pets.pop(3)
print(select1)
select2 = pets.pop()
print(select2)
print(pets)
````

####del 키워드
- del 리스트[인덱스]
- 리스트에서 지정한 인덱스의 항목을 삭제

````python
pets = ['dog', 'cat', 'hamster', 'hedgehog', 'parrot', 'lizard']
del pets[3], pets[-1]
print(pets)
````

####remove()
- 리스트.remove(값)

````python
pets = ['dog', 'cat', 'hamster', 'hedgehog', 'parrot', 'lizard']
#pets.remove(3)
pets.remove('hedgehog')
print(pets)
````

````python
pets = ['dog', 'cat', 'hamster', 'hedgehog', 'parrot', 'lizard']
pets.remove('cat')
print(pets)
````

#### clear()
- 리스트의 모든 항목을 삭제

````python
pets = ['dog', 'cat', 'hamster', 'hedgehog', 'parrot', 'lizard']
pets.clear()
print(pets)
````

###항목 재배열
- reverse()
- sort()

####reverse()
- 리스트 항목의 값은 그대로 두면서 순서만 재배열
- 리스트.reverse()

````python
nations = ['한국', '미국', '영국', '프랑스', '이탈리아']
nations.reverse()
print(nations)
````

####sort()
- 기본 오름차순으로 정렬
- 내림차순으로 정렬할 경우 reverse=True

````python
nations = ['한국', '미국', '영국', '프랑스', '이탈리아']
nations.sort()
print(nations)
nations.sort(reverse=True)
print(nations)
````

##리스트 탐색
- in 연산자: 리스트에 특정 값이 있는지를 인덱스로 알려줌
  - 값 in 리스트

##리스트와 랜덤

###랜덤 선택
- random 라이브러리의 choice()는 리스트의 항목 한개를 무작위로 선택
  - random.choice(리스트)
- random.sample() : 설정한 개수만큼 리스트에서 항목을 중복 없이 무작위로 선택
  - random.sample(리스트, 개수)
- random.choices(리스트, k=개수)
  - 리스트에서 항목 k개를 중복을 허용해 무작위로 선택

````python
import random
colors = ['ivory', 'lavender', 'lightcyan', 'lightyellow', 'palegreem']
print(random.choice(colors))
print(random.sample(colors,3))
print(random.choices(colors,k=3))
````
