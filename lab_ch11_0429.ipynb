###데이터 조작

####열 이름 변경하기
- rename()
  - 데이터프레임.rename(columns={'열이름':'새로운 열 이름'})
  - inplace = True 옵션을 추가하면 데이터 프레임에 변경된 열 이름이 바로 반영

````python
df.rename(columns = {'test':'exam'}, inplace = True)
display(df)
````

####데이터 정렬하기
- sort_values()
  - 데이터프레임.sort_values('열이름')

````python
sort_df = df.sort_values('sum', ascending=False)
display(sort_df)
````

````python
display(df)
````

####조건에 따라 다른 값 부여하기
- where() : 데이터 조건에 따라 값이 다른 열을 생성
- grade 열을 새롭게 생성
- sum 열의 값이 90 이상이면 A, 80 이상이면 B, 나머진 C

````python
import numpy as np

df['grade'] = np.where(df['sum']>=90, 'A', np.where(df['sum']>=80, 'B', 'C'))
display(df)
````

image.png

````python
df['result']=np.where(df['sum']>=80, 'pass','fail')
display(df)
````

###누락된 데이터 처리하기

####누락된 데이터가 포함된 데이터프레임 생성
- NaN(Not a Number)

````python
import numpy as np
import pandas as pd

df = pd.DataFrame({'name':['Jessica','Liam','Sophia','Ryan','Alex'],
                   'gender':['F',np.nan, np.nan, 'M','F'],
                   'score':[89,75,95,np.nan, 93]})
display(df)
````

####누락된 데이터 확인
- isna()
  - pd.isna(데이터프레임)
  - 누락된 데이터가 있는지 확인
  - 누락된 데이터는 True, 아닌 데이터는 False로 표시

````python
print(pd.isna(df))
````

- isna()와 sum()를 함께 사용하여 누락된 데이터 카운팅

````python
print(pd.isna(df).sum())
````

####누락된 데이터 삭제
-dropna()
  - 누락된 데이터가 포함된 행 또는 열 삭제
  - 기본적으로 행을 삭제하고 옵션 axis = 1 or axis = 'columns'를 추가하면 열을 삭제

````python
print(pd.isna(df))
````

````python
clean_df = df.dropna()
display(clean_df)
````

- 열이 누락된 행만 삭제하고자 하는 경우
  - subset 매개변수 값으로 지정
  
  image.png

````python
display(df)
clean_df=df.dropna(subset=['name','score'])
display(clean_df)
````

````python
df = pd.DataFrame({'name': ['Jessica', 'Liam', 'Sophia', 'Ryan', 'Alex'],
                   'gender': ['F', np.nan, np.nan, 'M', 'F'],
                   'score': [89, 75, 95, np.nan, 93]})
````

- 누락된 값을 삭제하여도 원본 데이터프레임에는 변화가 없음
- 원본 데이터프레임에도 내용을 반영하기위해서는 inplace = True 옵션을 추가

````python
display(df)
````

````python
clean_df = df.dropna(subset=['name','score'])
display(clean_df)
display(df)
clean_df = df.dropna(subset=['name','score'],inplace=True)
display(df)
````

####누락된 데이터를 다른 값으로 대체하기
- fillna()
  - 누락된 데이터를 값으로 대체
  - NaN을 0으로 대체하고, df1에 저장하는 코드
  - df 데이터프레임 원본에는 변화 없음

````python
df = pd.DataFrame({'name': ['Jessica','Liam','Sophia','Ryan','Alex'],
                   'gender' : ['F',np.nan,np.nan,'M','F'],
                   'score' : [89, 75, 95, np.nan, 93]})
df1 = df.fillna(0)
display(df1)
````

- gender 열의 NaN etc로 대체
- score 열의 NaN은 0으로 대체

````python
df1 = df.fillna({'gender':'etc','score':0})
display(df1)
````

- score 열의 누락된 데이터를 열 값의 평균으로 대체하려면 mean함수 사용

````python
df1 = df.fillna({'gender':'etc','score':df['score'].mean()})
display(df1)
````
