##딕셔너리의 개요
- 딕셔너리(dictionary) 리스트와 유사하게 하나 이상의 데이터를 저장하는 자료형
- 키(key)와 값(value)의 쌍으로 이루어짐
- 리스트는 인덱스로 접근하나, 딕셔너리는 값에 접근할 때는 키를 사용
- 대응 관계를 표현 (ex: 이름:홍길동, 생일:XX/XX/XX, ....)
- 연관배열(associative array) or 해시(hash)


###딕셔너리 생성
- 딕셔너리 이름 = {키1:값1,키2:값2, ...}

````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
print(prices)
````

###딕셔너리 키 조건
- 키 값이 동일한 경우 마지막 항목을 제외한 나머지 항목은 무시

````python
prices = {'아메리카노':2000, '라떼':3000, '아메리카노':1900, '녹차':2500}
print(prices)
````

###딕셔너리 값에 접근하기
- 딕셔너리 접근은 딕셔너리 이름을 쓰고 대괄호 안에 키 이름 표기

````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
print(prices['라떼'])
````

###딕셔너리 항목 추가
- 딕셔너리이름[키] = 값

````python
prices = {}
prices['아메리카노'] = 2000
prices['라떼'] = 3000
prices['녹차'] = 2500
print(prices)
````

###딕셔너리 값 수정
- 딕셔너리이름[키] = 값

````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
prices['아메리카노'] = 2500
print(prices)
````

##딕셔너리 다루기

###딕셔너리 순회하기
- 딕셔너리 키를 순회하는 방법


```
for 키 in 딕셔너리:
  문장
```



````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
for key in prices:
  print(key, end=' ')
````

````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
for key in prices:
  print(prices[key])
````

###함수로 딕셔너리 순회하기:items()
- items() : 키와 값을 동시에 얻는 방법
- 딕셔너리.items()
  - ```
for 키변수, 값변수 in 딕셔너리.items():
       문장
```





함수로 딕셔너리 순회하기 : key(), values()
- 딕셔너리 항목의 key()와 values()

````python
prices = {'아메리카노':2000, '라떼':3000, '녹차':2500}
for key, value in prices.items():
  print(key, value)
#for key in prices:
#  print(key)
#for values in prices.values():
#  print(values)
print(prices.keys())
print(prices.values())
````

##딕셔너리 항목 삭제하기

###pop함수로 항목 삭제하기
- 딕셔너리.pop(키)

````python
prices = {'아메리카노':2000,'라떼':3000,'녹차':2500}
a = prices.pop('라떼')
print(a)
print(prices)
````

###모든 항목 삭제
- clear()

````python
prices = {'아메리카노':2000,'라떼':3000,'녹차':2500}
a = prices.clear()
print(a)
print(prices)
````

###LAB 국가 수도 사전

````python
capitals= {'대한민국':'서울','미국':'워싱턴','일본':'도쿄'}
nation = input('국가 이름: ')
if nation in capitals:
  print(capitals[nation])
else:
  print(f'{nation}은 등록되지 않았습니다.')
````

###LAB 주문 금액 계산기
- 음료 종류와 주문 수를 입력받아 전체 주문 금액을 계산

````python
prices = {'아메리카노':2000,'라떼':3000,'녹차':2500}
sum = 0
while True:
  s = input('음료 종류 또는 종료: ')
  if s=='종료':
    break
  elif s in prices:
    n = int(input(f'{s} 주문 수: '))
    sum += prices[s]*n
  else:
    print(f'{s}는 없습니다.')
print(f'총 금액: {sum}원')
````

###혈액형 빈도수

````python
blood = {'A':0,'B':0,'O':0,'AB':0}
````

````python
while True:
  s = input('혈액형(A, B, O, AB) 또는 종료: ')
  if s=='종료':
    print('종료합니다.')
    break
  elif s in blood:
    blood[s] += 1
  else:
    print('잘못 입력했습니다.')
````

````python
for key, value in blood.items():
  print(key, value)
````
